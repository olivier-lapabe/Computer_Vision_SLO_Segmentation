```python
import numpy as np
from scipy.ndimage import rotate
```

```python
def rect(width, height, rotation=0):
    # Create a rectangle with ones
    rectangle = np.ones((height, width))
    # Rotate the rectangle if needed
    if rotation != 0:
        # Increase size before rotation to avoid clipping
        larger_side = max(width, height) * 2
        larger_rectangle = np.zeros((larger_side, larger_side))
        start_x = (larger_side - height) // 2
        start_y = (larger_side - width) // 2
        larger_rectangle[start_x:start_x+height, start_y:start_y+width] = rectangle
        # Rotate and return
        rotated = rotate(larger_rectangle, rotation, reshape=False, mode='constant', cval=0)
        return rotated
    return rectangle
```

```python
import itertools

# Define ranges for rectangle parameters
widths = [1, 2, 3]  # Example widths
heights = [1, 2, 3]  # Example heights
orientations = [0, 45, 90]  # Example orientations

# Function to generate all combinations of a single rectangle
def generate_single_rect_combinations(widths, heights, orientations):
    return itertools.product(widths, heights, orientations)

# Function to generate p_structures with varying numbers of rectangles
def generate_p_structures(widths, heights, orientations, max_rectangles=2):
    p_structures = []
    single_rect_combinations = list(generate_single_rect_combinations(widths, heights, orientations))
    
    for r in range(1, max_rectangles + 1):
        # Generate all combinations of r rectangles
        for combination in itertools.combinations(single_rect_combinations, r):
            # Convert each combination to the desired format
            p_structure = [rect(w,h,o) for w, h, o in combination]
            p_structures.append(p_structure)
    
    return p_structures

# Generate p_structures
p_structures = generate_p_structures(widths, heights, orientations)

# Example: print the first 5 p_structures to check
i=0
for p in p_structures:
    print(p)
    i += 1
```
